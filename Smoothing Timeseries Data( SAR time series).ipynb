# SavGol Filter

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = 'Desktop/ee-chart.csv'
```

```python
df=pd.read_csv(data)
df.head()
```

```python
df.plot()
plt.title('Raw VV')
```

```python
from scipy.signal import savgol_filter
```

```python
y = df['VV'].to_numpy() #Converting pandas dataframe to numpy array
```

```python
print(y.shape)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.plot(y)
plt.title('Raw VV')
plt.subplot(222)
plt.plot(savgol_filter(y,11,5))
plt.title('Smoothen VV')
```

# Convolution Filter

```python
def smooth(y,box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y,box,mode='same')
    return y_smooth
```

```python
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.plot(y)
plt.title('Raw VV')
plt.subplot(222)
plt.plot(smooth(y,2))
plt.title('Smoothen VV')
```
