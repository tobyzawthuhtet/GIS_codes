```python
from PIL import Image
from numpy import asarray
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = asarray(Image.open(r'C:\Users\Toby\Desktop\Aregentia_R.tif'))
print('Shape :',image.shape)
print('Unique image :',np.unique(image))
```

```python
print("Minimum",np.nanmin(image))
print("Max",np.nanmax(image))
print("Mean",np.nanmean(image))
```

```python
import gdal
raster = gdal.Open(r'C:\Users\Toby\Desktop\Aregentia_R.tif')
srcband = raster.GetRasterBand(1)

# Get raster statistics
stats = srcband.GetStatistics(True, True)

print("Minimum",stats[0])
print("Max",stats[1])
print("Mean",stats[2])
```
